```python
import polars as pl
import seaborn as sns
import matplotlib.pyplot as plt
print("Imported libs")
```

```python
full_df = pl.read_csv("all_combined.csv", null_values=["null"])
full_df = full_df.with_columns(
    pl.from_epoch("timestamp", time_unit="s")
)
full_df
```

```python
grouped_by_prefix = (
    full_df
    .with_columns(
        prefix=pl.col("balancerId").str.slice(0, pl.col("balancerId").str.find("f")).cast(pl.Int32)
    )
    .group_by("prefix")
    .agg(
        pl.col("balancerColocationCenter").unique().alias("colocationCenters"),
        pl.col("balancerId").unique().alias("balancerIds"),
        pl.col("balancerId").n_unique().alias("balancerCount"),
    )
).sort("balancerCount")

grouped_by_prefix
```

```python
grouped_by_prefix.filter(pl.col("colocationCenters").list.len() > 1)
```

Empty, indicating prefixes are 1:1 with colocation centers

```python
ids_without_prefix = grouped_by_prefix.with_columns(
    pl.col("balancerIds").list.eval(
        pl.element().str.split("f").list.get(1).cast(pl.Int32)
    )
)

ids_without_prefix = ids_without_prefix.with_columns(
    pl.col("balancerIds").list.sort()
)


with_max_id = ids_without_prefix.with_columns(
    pl.col("balancerIds").list.max().alias("max_id")
    )


with_max_id
```

If the assumption can be made that the balancer IDs are increasing integers, then the max_id seen indicates there are at least that many instances for that prefix.

```python
max_is_count = with_max_id.filter(pl.col("max_id") == pl.col("balancerCount"))

max_is_count
```

These are the ones where the max id matches the number of balancer seen. Assuming there is no other max, can somewhat confidently say we've seen all instances of that prefix

```python
grouped_by_prefix.sort("prefix")["prefix"].to_list()
```

Prefixes are just incremented numbers, does this mean anything? Does a smaller number indicate an older 'group'? If so, this an indicate the evolution of a colocation center. Additionally, can missing numbers in the middle indicate a blind spot in our data

```python
grouped_by_colocation = (
    full_df
    .with_columns(
        prefix=pl.col("balancerId").str.slice(0, pl.col("balancerId").str.find("f")).cast(pl.Int32)
    )
    .group_by("balancerColocationCenter")
    .agg(
        pl.col("prefix").unique().alias("prefixes"),
        pl.col("prefix").n_unique().alias("prefixCount"),
        pl.col("balancerId").unique().alias("balancerIds"),
        pl.col("balancerId").n_unique().alias("balancerCount"),
    )
)
grouped_by_colocation
```
