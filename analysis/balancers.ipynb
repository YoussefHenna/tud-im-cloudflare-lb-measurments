```python
import polars as pl
import seaborn as sns
import matplotlib.pyplot as plt
print("Imported libs")
```

```python
df = pl.read_csv("regional_dataset_combined.csv")
df = df.with_columns(
    pl.from_epoch("timestamp", time_unit="s")
)
df
```

### Client
Client -- single vantage point, defined by city and asn.

```python
client_cols = ['clientCountry', 'clientCity', 'clientAsn', 'clientNetwork']
requests_by_client = df.group_by(client_cols).agg(
    pl.len().alias('requests')
).sort('requests')
requests_by_client
```

### Balancer colocation center
Balancer colocation center is singe datacenter.
Let's check how many colocations we have for each client.

```python
colocations_by_client = df.group_by(client_cols).agg(
    pl.col('balancerColocationCenter').unique().len().alias('colocations')
).sort('colocations', descending=True)
colocations_by_client
```

Let's check how distributed requests where one client is routed to different colocations.

```python
requests_distribution = df.filter(
    pl.col('balancerColocationCenter').n_unique().over(client_cols) > 1
).group_by(client_cols + ['balancerColocationCenter']).agg(
    pl.len().alias('requests')
).sort(client_cols + ['requests'], descending=[False] * len(client_cols) + [True])
print(requests_distribution)
```

```python
not_full_countries = df.group_by('clientCountry').agg(pl.len().alias('requests')).filter(pl.col('requests') < 3000)
not_full_countries
```

```python
buffalo_requests = df.filter(
    (pl.col('clientCountry') == 'US') & 
    (pl.col('clientCity') == 'Buffalo') & 
    (pl.col('clientAsn') == 36352)
)
buffalo_requests
```

```python
yyz_data = buffalo_requests.filter(pl.col("balancerColocationCenter") == "YYZ")["latencyTotal"]
buf_data = buffalo_requests.filter(pl.col("balancerColocationCenter") == "BUF")["latencyTotal"]

# 2. Initialize the plot
# We use a shared bin range so the comparison is accurate
# You can adjust bins based on the range of your latency data
bins = 50 

plt.hist(yyz_data, bins=bins, alpha=0.5, range=(0, 200), label='YYZ', color='blue', edgecolor='none')
plt.hist(buf_data, bins=bins, alpha=0.5, range=(0, 200), label='BUF', color='orange', edgecolor='none')

# 3. Formatting the plot
plt.xlabel('Latency (ms)')
plt.ylabel('Requests')
plt.title('Latency distribution of requests to YYZ and BUF colos from buffalo')
plt.legend(loc='upper right')

# Add a grid for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Ensure layout is tight so labels aren't cut off
plt.tight_layout()

# Show the plot
plt.show()
```
